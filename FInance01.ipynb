```python
#libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Simple Interest

```python
PV = 1000
i = 0.10 #10% per annum
n = 100 # years
#FV = ?

FVS = {'Present Value':[], 'Interest % per annum':[], 'Year':[], 'Future Value':[]}

for x in range(0,n+1):
    value = PV*(1+i*x)
    y = round(value, 2)
    FVS['Present Value'].append(PV)
    FVS['Interest % per annum'].append(i*100)
    FVS['Year'].append(x)
    FVS['Future Value'].append(y)
SimpleInterest = pd.DataFrame(data=FVS, index=None)
print("Simple Interest Table")
SimpleInterest
```

```python
yaxis = SimpleInterest['Future Value']
xaxis = SimpleInterest['Year']
fig, ax = plt.subplots(figsize=(12,8), constrained_layout=True)
ax.plot(xaxis, yaxis)
ax.set_xlabel('Years')
ax.set_ylabel('Future Value')
ax.set_title('Simple Interest')
fig.set_facecolor('lightsteelblue')
```

# Compounding Interest

```python
PV = 1000
i = 0.10 #10% per annum
n = 100 # years
#FV = ?

FVS2 = {'Present Value':[], 'Interest % per annum':[], 'Year':[], 'Future Value':[]}

for x in range(0,n+1):
    value2 = PV*(1+i)**x
    z = round(value2, 2)
    FVS2['Present Value'].append(PV)
    FVS2['Interest % per annum'].append(i*100)
    FVS2['Year'].append(x)
    FVS2['Future Value'].append(z)
CompoundingInterest = pd.DataFrame(data=FVS2, index=None)
print("Compounding Interest Table")
CompoundingInterest
```

```python
yaxis2 = CompoundingInterest['Future Value']
xaxis2 = CompoundingInterest['Year']
fig, ax = plt.subplots(figsize=(12,8), constrained_layout=True)
ax.plot(xaxis2, yaxis2)
ax.set_xlabel('Years')
ax.set_ylabel('Future Value')
ax.set_title('Compounding Interest')
fig.set_facecolor('lightsteelblue')
```
